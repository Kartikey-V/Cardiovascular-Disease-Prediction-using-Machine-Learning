## Splitting Dataset into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

Y = df['target']
X = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
feature_cols = ['age', 'sex', 'cp', 'trestbps', 'chol', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=26)
print("Test samples :", X_test.shape[0])
print("Training samples:",X_train.shape[0])
```

## Model Develpoment and Evalaution

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
from sklearn import model_selection
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
predicted = cross_val_predict(lr, X_test, Y_test, cv=10)
kfold = model_selection.KFold()
lr.fit(X_train,Y_train)
scores = model_selection.cross_val_score(lr, X_test, Y_test, cv=10)
print('Accuracy = ',(scores.mean()*100, scores.std()*100))
print("Precision:",metrics.precision_score(Y_test, predicted))
print("Recall:",metrics.recall_score(Y_test, predicted))
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(Y_test, predicted).ravel()
print(tn, fp, fn, tp)
confusion_matrix = [[tn, fp], [fn, tp]]
```

```python
sensitivity = (tp)/(tp+fn)
print('Sensitivity : ', sensitivity )
specificity = (tn)/(tn+fp)
print('Specificity : ', specificity)
```

```python
class_names=[0,1] 
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(confusion_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix for Logistic Regression', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
Y_pred_proba = lr.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(Y_test,  Y_pred_proba)
auc = metrics.roc_auc_score(Y_test, Y_pred_proba)
print("AUC = ",auc)
lw = 2
plt.plot(fpr,tpr,color='red',label="Logistic Regression AUC="+str(auc))
plt.plot([0, 1], [0, 1], color='cyan', lw=lw, linestyle='--')
plt.legend(loc=4)
plt.title('ROC curve for Logistic Regression')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
plt.show()
```
